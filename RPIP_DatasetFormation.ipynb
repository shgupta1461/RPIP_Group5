```python
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RPIP/movies.dat', delimiter='\t', encoding='latin-1')

# Create a DataFrame
movies = pd.DataFrame(movies)
movies = movies[['id', 'title', 'imdbID']]
movies.head()
```

```python
user_ratedmovies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RPIP/user_ratedmovies-timestamps.dat', delimiter='\t', encoding='utf-8')

# Create a DataFrame
user_ratedmovies = pd.DataFrame(user_ratedmovies)
user_ratedmovies = user_ratedmovies[['userID', 'movieID', 'rating']]
user_ratedmovies.head()
```

```python
df3 = pd.merge(user_ratedmovies, movies, left_on='movieID', right_on='id', how='inner')

# Drop unnecessary columns and rename columns
df3 = df3.drop(columns=['id'])
# Display df3
df3.head(15)
```

```python
df3 = df3[df3['rating'] > 4.0]
len(df3)
```

```python
def get_movie_name_by_id(df, movie_id):
    movie_name = df.loc[df['movieID'] == movie_id, 'title'].iloc[0]
    return movie_name
```

```python
#Example run
n = get_movie_name_by_id(df3, 1)
n
```

```python
unique_movie_names = df3['title'].unique().tolist()
print(unique_movie_names)
```

```python
# We have taken only 2647 movies into consideration so there is a keyboard interrupt
keywords = []
count = 0
for movie_name in unique_movie_names:
    try:
        k = imdb.get_by_name(movie_name)['keywords']
    except KeyError:
        k = 0  # Assign 0 when 'keywords' key is not found
    keywords.append(k)
    count += 1
    print(count)
```

```python
movie_keywords = {'movieName': unique_movie_names[0:2647], 'keywords': keywords}
movie_keywords = pd.DataFrame(movie_keywords)
movie_keywords.to_csv('/content/drive/MyDrive/Colab Notebooks/RPIP/movieKeywords.csv', index=False)
movie_keywords.head()
```

```python
d = pd.merge(df3, movie_keywords, left_on='title', right_on='movieName', how='left')
d.drop(columns=['movieName'], inplace=True)
d.head(20)
```

```python
d.to_csv('/content/drive/MyDrive/Colab Notebooks/RPIP/finalDataset.csv', index=False)
```

```python
filtered_df = d[d['keywords'].astype(bool)]
print(filtered_df)
```

```python
filtered_df = filtered_df.dropna(subset=['keywords'])
filtered_df = filtered_df[filtered_df['keywords'].astype(bool)]

# Display filtered DataFrame
print(filtered_df)
```
